### Locations

```python
def find_locations(doc):
    df = pd.DataFrame(
    [[x.text, x.start, x.end, x.sent] for x in doc.ents if x.label_ is 'LOC'],
    columns=['Location', 'Start', 'End','Sentence']
             )
    df['NZGazetteer']=[True if X in nzgazAntartica else False for X in df.Location ]
    df['ScarNZ']=[True if X in SCARNZnames else False for X in df.Location ]
    df['ScarGlobal']=[True if X in SCARGlobalnames else False for X in df.Location ]
    
    return df
```

### Finding Locations in Collection of Documents

Running through the folder of converted text documents we have so far to see what sort of extracted locations *are not* confirmed by either of the two gazetteers to see what refinement needs to be added. More gazetteers, and some refienment where a name begins with 'the' are most likely next steps.

```python
def findLocationsInAllDocs(folder, NLPmodel='large'):
    
        
    filelist = [x for x in os.listdir(folder) if x.endswith('txt')]
    df = pd.DataFrame()
    for filename in filelist:
        file = open(folder+'/'+ filename)
        doc = file.read()
        file.close()
        if NLPmodel == 'small':
            document = nlp(doc)
        elif NLPmodel == 'medium':
            document = nlp_md(doc)
        elif NLPmodel == 'large':
            document = nlp_lg(doc)
    
        df2 = find_locations(document)
        df2['Source'] = filename
        df = pd.concat([df,df2])
        df.reset_index(drop=True)
        
    return df
```

### Code

```python
from spacy.matcher import Matcher
```

```python
def findSpecies(doc):
    
    """"
    Dictionary approach.
    
    This code runs through token by token looking to match Genus + species, 
    Genus + species + species, G. species etc based on words found in the genus 
    and species columns of the supplied list of species.
    
    Also finds examples where a known genus is followed by an unreginsied species 
    if that example also appears in an abbreviated form elsewhere. 
    """"

    foundSpeciesList = []
    foundSpecies = ''
    foundSpeciestokens = []
    foundSpeciesDict ={}
    possiblesNames =[]
    possiblesTokens = []

    
    for token in doc:
        if str(token) in uniqueGenus:
            foundSpecies = foundSpecies + str(token)
            foundSpeciestokens.append(token)
            j=1
            onlyGenus = True
            while (doc[token.i+j].text in uniqueSpecies):
                foundSpecies += ' ' + str(doc[token.i+j])
                foundSpeciestokens.append(doc[token.i+j])
                j+=1
                onlyGenus = False
                
            key = ''
            
            if onlyGenus ==False:
                for tkn in foundSpeciestokens[:-1]:
                    key += tkn.text[0] + '. '
                key+= foundSpeciestokens[-1].text
                if key not in foundSpeciesDict:
                    foundSpeciesDict[key]=[foundSpecies, True]
#                     print('adding ', key, ': [', foundSpecies, True,'] to dict' )

                foundSpeciesList.append([foundSpecies, foundSpecies, True, token.i, token.i+j, token.idx, token.sent])
#                 print('DEBUG 1 - adding ',foundSpecies, foundSpecies, True, token.i, 'to DF')

                
            else:
                candidate = token.text + ' ' + doc[token.i+1].text
                if candidate not in possiblesNames:
                    possiblesNames.append(candidate)
                    possiblesTokens.append((token, doc[token.i+1]))

            foundSpecies=''
            foundSpeciestokens=[]


        elif token.shape_ == 'X.':
            foundSpeciestokens=[]
            foundSpecies = foundSpecies + str(token)
            foundSpeciestokens.append(token)
            j=1
            while (doc[token.i+j].text in uniqueSpecies or doc[token.i+j].shape_ == 'x.'):
                foundSpecies += ' ' + str(doc[token.i+j])
                foundSpeciestokens.append(doc[token.i+j])
                j+=1
            if doc[token.i+j-1].text in uniqueSpecies:
#                 print(doc[token.i+j-1].text, 'in uniqueSpecies')
                if foundSpecies in foundSpeciesDict:
                    foundSpeciesList.append([foundSpecies, foundSpeciesDict.get(foundSpecies)[0], foundSpeciesDict.get(foundSpecies)[1], token.i, token.i+j, token.idx, token.sent])
#                     print('DEBUG 2 -adding ',foundSpecies, foundSpeciesDict.get(foundSpecies)[0], foundSpeciesDict.get(foundSpecies)[1],token.i, 'to DF')
                else:
#                     print("looking for", foundSpeciestokens)
                    matcher = Matcher(nlp.vocab)
                    pattern = [{"SHAPE": "Xxxxx", 
                                   "TEXT"[0]: foundSpeciestokens[0].text[0]}
                              ]
                    
                    for tkn in foundSpeciestokens[1:-1]:
                        pattern.append({"SHAPE": "xxxx", 
                                       "TEXT"[0]: tkn[0].text[0]})
                    pattern.append({"TEXT": foundSpeciestokens[-1].text})
#                     print("looking for", pattern)                
                    matcher.add("Species_Pattern", None, pattern)

                    matches = matcher(doc)
                    if len(matches)>0:
                        start = matches[0][1]
                        end =  matches[0][2]
                        matched_span = doc[start:end]
                        inList = False
                        if doc[start].text in uniqueGenus:
                            inList = True
#                             print(doc[start].text, 'in UniqueGenus')

                        foundSpeciesList.append([foundSpecies, matched_span.text, inList, token.i, token.i+j, token.idx, token.sent])
#                         print('DEBUG 3 -adding ',foundSpecies, matched_span.text, inList, token.i, 'to DF')

                        foundSpeciesDict[foundSpecies]=[matched_span.text, inList]

                        if inList==False:
                            for match_id, start, end in matches:
                                matched_span = doc[start:end]
                                foundSpeciesList.append([matched_span.text, matched_span.text, inList, start,end, doc[start].idx, doc[start].sent])
#                                 print('DEBUG 4 -adding ',matched_span.text, matched_span.text, inList, start, 'to DF')
            foundSpecies=''
            foundSpeciestokens=[]
                
#     print(possiblesTokens)
    for tkns in possiblesTokens:
        shortPattern = [{"SHAPE": "X.", "TEXT"[0]: tkns[0].text[0]},
                           {"TEXT": doc[tkns[0].i+1].text}
                          ]
        fullPattern = [{"SHAPE": "Xxxxx", "TEXT": tkns[0].text},
                           {"TEXT": doc[tkns[0].i+1].text}
                          ]
#         print('candidate = ', tkns)
#         print("looking for", shortPattern) 
        matcher = Matcher(nlp.vocab)
        matcher.add("abbreviated_Species_Pattern", None, shortPattern)
        matches = matcher(doc)
        if len(matches)>0:
#             print('found abbreviated version of candidate')
            matcher.add("full_Species_Pattern", None, fullPattern)
            
            matches = matcher(doc)
            for match_id, start, end in matches:
                matched_span = doc[start:end]
                foundSpeciesList.append([matched_span.text, tkns[0].text + ' '+ tkns[1].text, False, start, end, doc[start].idx, doc[start].sent])
#                 print('DEBUG 5 -adding ',matched_span.text, tkns[0].text + ' '+ tkns[1].text, False ,start, 'to DF')



    FoundSpeciesDF = pd.DataFrame(foundSpeciesList,columns=['Found as', 'Full name', 'InDefinitiveList', 'Start', 'End','PositionOfFirstToken','Sentence'])
    FoundSpeciesDF.sort_values('Start', inplace=True)
    FoundSpeciesDF.reset_index(inplace=True,drop=True)
#     FoundSpeciesDF['InReferenceList']=FoundSpeciesDF['Full name']
    return FoundSpeciesDF
```
